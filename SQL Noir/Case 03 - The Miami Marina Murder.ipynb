## Case #003: The Miami Marina Murder

A body was found floating near the docks of Coral Bay Marina in the early hours of August 14, 1986. Your job, detective, is to find the murderer and bring them to justice. This case might require the use of JOINs, wildcard searches, and logical deduction. Get to work, detective.

### Objectives

1. Find the murderer. ( Start by finding the crime scene and go from there )

<img src="./images/Screenshot 2026-02-05 135317.png">

- Same as the previous cases, we need to get all the details about the current case

```python
SELECT * 
FROM crime_scene
WHERE location = 'Coral Bay Marina'
    AND date = 19860814
```

Description: 

The body of an unidentified man was found near the docks. Two people were seen nearby: one who lives on 300ish "Ocean Drive" and another whose first name ends with "ul" and his last name ends with "ez".

```python
SELECT *
FROM person
WHERE address LIKE "%3__ Ocean Drive%"
OR name LIKE '%ul %ez'
```

<img src="./images/Screenshot 2026-02-05 140429.png">

- Using the previous code as a base, we can create a query to get their testimonies. (We can easily use the fetched ids from the previous execution, but where's the fun in that)

```python
SELECT p.name, i.transcript
FROM person AS p
JOIN interviews AS i
    ON p.id = i.person_id
WHERE p.address LIKE "%3__ Ocean Drive%"
    OR p.name LIKE '%ul %ez'
```

<img src="./images/Screenshot 2026-02-05 141258.png">

```python
SELECT p.name, sr.suspicious_activity
FROM person AS p
JOIN hotel_checkins AS hc
    ON p.id = hc.person_id
JOIN surveillance_records AS sr
    ON p.id = sr.person_id
WHERE hc.hotel_name LIKE '%Sunset%'
    AND hc.check_in_date LIKE '____0813'
    AND sr.suspicious_activity IS NOT NULL
```

- The query above uses the descriptions provided by the witnesses. However, the result does not seem to narrow down the list of names due to the number of entries yielded. Thus, we need to filter the `suspicious activity`. Based on the results of the query, the following are suspcious enough for me:

- `Spotted entering late at night`
- `Left suddenly at 3 AM`
- `Seen arguing with an unknown person`

Using these, let's get their interviews 

```python
SELECT p.name, i.transcript
FROM person AS p
JOIN interviews AS i
    ON p.id = i.person_id
JOIN surveillance_records AS sr
    ON p.id = sr.person_id
WHERE sr.suspicious_activity IN (
    'Spotted entering late at night',
    'Left suddenly at 3 AM',
    'Seen arguing with an unknown person'
)
```

<img src="./images/Screenshot 2026-02-05 161906.png">

- Apparently, this does not help at all. The `interview` table was supposed to be for the witness. Let's try applying the previous code on the `confessions` table

```python
SELECT p.name, c.confession
FROM person AS p
JOIN confessions AS c
    ON p.id = c.person_id
JOIN surveillance_records AS sr
    ON p.id = sr.person_id
WHERE sr.suspicious_activity IN (
    'Spotted entering late at night',
    'Left suddenly at 3 AM',
    'Seen arguing with an unknown person'
)
```
